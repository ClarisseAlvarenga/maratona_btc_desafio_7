```
import paho.mqtt.client as mqtt
from csv import writer
import json

dados_sem_titulo = ['Tempo','Estação','LAT','LONG',
                    'Movimentação','Original_473',
                    'Original_269','Zero','Maçã-Verde',
                    'Tangerina','Citrus','Açaí-Guaraná',
                    'Pêssego','Status','row']
append_list_as_row('desafio7.csv',dados_sem_titulo)

```

```
# The callback for when the client receives a CONNACK response from the server.
def on_connect(client, userdata, flags, rc):
    print("Connected with result code "+str(rc))
    # Subscribing in on_connect() means that if we lose the connection and
    # reconnect then subscriptions will be renewed.
    client.subscribe("tnt")

# The callback for when a PUBLISH message is received from the server.
def on_message(client, userdata, msg):
    print(msg.topic+" "+str(msg.payload))
    dados_zoados = msg.payload
    #decoding data in a readable manner
    dict_dados = json.loads(dado_zoados.decode('utf-8'))
    #usando o for como list comprehension
    dados_sem_titulo = [value for key , value in dict_dados.items() if value is not None]
    append_list_as_row('desafio7.csv',dados_sem_titulo)
    if len(dados_sem_titulo) >= 20000:
      client.disconnect()

#function to write csv
def append_list_as_row(file_name, list_of_elem):
    #Open file in append mode
    with open(file_name, 'a+', newline='') as write_obj:
        # Create a writer object from csv module
        csv_writer = writer(write_obj)
        # Add contents of list as last row in the csv file
        csv_writer.writerow(list_of_elem)



#creating a connection
client = mqtt.Client()
client.on_connect = on_connect
client.on_message = on_message

client.username_pw_set("maratoners","ndsjknvkdnvjsbvj")

client.connect("tnt-iot.maratona.dev", 30573, 60)

# Blocking call that processes network traffic, dispatches callbacks and
# handles reconnecting.
# Other loop*() functions are available that give a threaded interface and a
# manual interface.
#creating a condition to stop capturing data
client.loop_forever()
```
